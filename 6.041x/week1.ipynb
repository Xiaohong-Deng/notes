Probability Axioms

Infinite Continuous Sample space中, 一个element的概率是0, 以单位矩形Unit Square为例, 矩形内一点概率为0. 一般对这样的space取subset, 概率方为正, 这样的subset叫event

Axioms

- Nonnegativity P(A) > 1
- Normalization $P(\Omega) = 1$
- Additivity If A^B = empty, then P(A+B) = P(A) + P(B);  A+B这里指并集, A^B交集

P(A+B) = P(A) + P(B) - P(A^B) => P(A+B) <= P(A) + P(B)  这就是Union Bound, P(A+B)比右侧小多少取决于P(A^B)的大小咯, 右侧的值是可能超过1的, 除非A, B没有交集
P(A+B+C)=P(A) + P(B) + P(C) - P(A^B) - P((A+B)^C) 看不出event增加会增加左侧概率还是减少
A+B+C=A + (B^(A's complement)) + (C^A's comp^B's comp), 右侧全disjoint, so P(A+B+C) = P(A) + P(A's comp^B) + P(A's comp ^ B's comp ^ C)
